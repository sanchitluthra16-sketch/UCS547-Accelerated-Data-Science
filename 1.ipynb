```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
v, c = np.unique(x, return_counts=True)
m = v[c == c.max()]
i = [np.where(x == k)[0] for k in m]
print(m, i)
```

```python
y = np.array([1,1,1,2,3,4,2,4,3,3])
v, c = np.unique(y, return_counts=True)
m = v[c == c.max()]
i = [np.where(y == k)[0] for k in m]
print(m, i)
```
